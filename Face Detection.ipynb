```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 # open-cv
```

```python
img=plt.imread("groupimage.jpg")
img
```

```python
plt.imshow(img)
```

```python
model=cv2.CascadeClassifier("haarcascade_frontalface.xml")
```

```python
model
```

```python
faces=model.detectMultiScale(img)
faces
```

```python
x,y,w,h = faces[0]
```

```python
img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0))
```

```python
img
```

```python
plt.imshow(img)
```

```python
for one_face in faces:
    x,y,w,h=one_face
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)
```

```python
plt.imshow(img)
```
